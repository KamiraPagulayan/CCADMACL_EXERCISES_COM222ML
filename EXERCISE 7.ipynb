```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.frequent_patterns import fpgrowth
```

```python
#reading data
url = "https://github.com/robitussin/CCADMACL_EXERCISES/blob/ff41fec98bd270f4bbeedb9b8849ae21fdeadfee/Groceries_dataset.csv?raw=true"
df = pd.read_csv(url)

df.head()
```

```python
client_basket = df.groupby(['Member_number', 'Date'])['itemDescription'].apply(lambda x: ', '.join(x))
client_basket
```

### 1. Apply one-hot encoding

```python
def encode_units(x):
    if x <= 0:
        return False
    if x >= 1:
        return True

basket_sets = client_basket.str.get_dummies(sep=', ')
basket_sets = basket_sets.applymap(encode_units)
basket_sets.head()
```

```python
one_hot_transformer = TransactionEncoder()
df_transform = one_hot_transformer.fit_transform(df)
df_transform
```

```python
df = pd.DataFrame(df_transform,columns=one_hot_transformer.columns_)
df
```

2. Use the Apriori Algorithm to find the most frequent sets

```python
frequent_itemsets = apriori(basket_sets, min_support=0.005, use_colnames=True)
print(frequent_itemsets)
```

3. Display the association rules using the lift metric

```python
df_ar = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```
